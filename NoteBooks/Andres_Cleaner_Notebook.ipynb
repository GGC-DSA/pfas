## UCMR3 Dataset

### Breaking Down to Region and into only PFAS chemicals

```python
ucmr3_df_regions = ucmr3_df[ucmr3_df['Region'] == 4]
pfas_ucmr = ucmr3_df_regions[ucmr3_df_regions['Contaminant'].str.startswith('PF')]
pfas_ucmr
```

```python
pfas_confirmed = pfas_ucmr.dropna(axis=0)
pfas_confirmed
```

### Georgia PFAS

Keeping the FacilityName, Associated FacilityID to possibly also use to match with the facotries

```python
pfas_ga = pfas_confirmed[pfas_confirmed['State'] == 'GA']
pfas_ga = pfas_ga.drop(['FacilityID', 'Region', 'State', 'SamplePointID', 'SamplePointType', 'Size', 'AssociatedSamplePointID', 'SampleID', 'MethodID', 'SampleEventCode', 'MonitoringRequirement', 'SamplePointName', 'FacilityWaterType'], axis = 1)
ids = pfas_ga['PWSID'].unique().tolist()
pfas_ga
```

In the graph below I was seeing the found contamination in each of the counties in Georgia. This would allow me to see if there is a growth in contamination in people in the future. Whenever I get help to open the dataset though

```python
pfas_ga['PWSName'].hist(bins = 15, xrot = 90)
plt.title('Amount of PFAS Contamination per County')
plt.ylabel('Recorded Contamination')
```

This one requires a tiny bit of explanation. MRL has shown in the graph also means minimum reporting level for a certain chemical. What that means is that if the result meets or succeeeds this, it is consider contaminated and is required to be reported. Furthermore their are several variations of PFAS so MRI can be different for each. Now the analyzed result value is simply the result from the testing in the water. Now I am comparing the two see how the result value comapred to mrl in this dataset.

```python
pfas_ga.plot.scatter('MRL', 'AnalyticalResultValue')
plt.title('MRL vs Analyzed Result Value')
plt.ylabel('Analytical Result Value')
plt.xlabel('Minimum Reporting Level')
```
